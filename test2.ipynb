```python
import xml.etree.ElementTree as ET
import os


xml_path = "go_cart_anno.xml"
output_dir = "go_cart_labels"
class_mapping = {"person": 0, "go_cart": 1}  
image_width = 852  
image_height = 480  


os.makedirs(output_dir, exist_ok=True)


tree = ET.parse(xml_path)
root = tree.getroot()


annotations = {}


for track in root.findall("track"):
    label = track.get("label")
    class_id = class_mapping.get(label)
    
    if class_id is None:
        continue  

    for box in track.findall("box"):
        frame = int(box.get("frame"))
        xtl = float(box.get("xtl"))
        ytl = float(box.get("ytl"))
        xbr = float(box.get("xbr"))
        ybr = float(box.get("ybr"))
        
        
        x_center = (xtl + xbr) / (2 * image_width)
        y_center = (ytl + ybr) / (2 * image_height)
        width = (xbr - xtl) / image_width
        height = (ybr - ytl) / image_height
        
        
        yolo_line = f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}"
        
        if frame not in annotations:
            annotations[frame] = []
        annotations[frame].append(yolo_line)


for frame, lines in annotations.items():
    filename = f"frame_{frame:06d}.txt"
    with open(os.path.join(output_dir, filename), "w") as f:
        f.write("\n".join(lines))

print(f"Конвертация завершена! Сохранено {len(annotations)} аннотаций в {output_dir}/")
```

```python
import os
import shutil
import random
from sklearn.model_selection import train_test_split

# Конфигурация
SOURCE_IMAGES = "dataset/go_cart_frames"         # Исходные кадры
SOURCE_LABELS = "dataset/go_cart_labels"         # Исходные аннотации
TRAIN_IMAGES = "dataset/train/go_cart_images"    # Train изображения
TRAIN_LABELS = "dataset/train/go_cart_labels"    # Train аннотации
VAL_IMAGES = "dataset/val/go_cart_images"        # Val изображения
VAL_LABELS = "dataset/val/go_cart_labels"        # Val аннотации
TEST_SIZE = 0.15                          # Доля val данных 
RANDOM_SEED = 42                         # Для воспроизводимости

# Создаем целевые папки
os.makedirs(TRAIN_IMAGES, exist_ok=True)
os.makedirs(TRAIN_LABELS, exist_ok=True)
os.makedirs(VAL_IMAGES, exist_ok=True)
os.makedirs(VAL_LABELS, exist_ok=True)

# Получаем список базовых имен файлов (без расширения)
all_files = [f.split(".")[0] for f in os.listdir(SOURCE_IMAGES) if f.endswith(".jpg")]
all_files.sort()

# Разделяем на train/val
train_files, val_files = train_test_split(
    all_files, 
    test_size=TEST_SIZE, 
    random_state=RANDOM_SEED
)

# Функция для копирования файлов
def copy_files(files, src_img, src_lbl, dest_img, dest_lbl):
    for base in files:
        # Копируем изображения
        shutil.copy(
            os.path.join(src_img, base + ".jpg"),
            os.path.join(dest_img, base + ".jpg")
        )
        # Копируем аннотации
        lbl_file = base + ".txt"
        if os.path.exists(os.path.join(src_lbl, lbl_file)):
            shutil.copy(
                os.path.join(src_lbl, lbl_file),
                os.path.join(dest_lbl, lbl_file)
            )

# Копируем данные
copy_files(train_files, SOURCE_IMAGES, SOURCE_LABELS, TRAIN_IMAGES, TRAIN_LABELS)
copy_files(val_files, SOURCE_IMAGES, SOURCE_LABELS, VAL_IMAGES, VAL_LABELS)

print(f"Разделение завершено!")
print(f"Всего файлов: {len(all_files)}")
print(f"Train: {len(train_files)}")
print(f"Val: {len(val_files)}")
```

```python
import cv2
import os

def plot_yolo_boxes(image_path, label_path):
    # Проверяем существование файлов
    if not os.path.exists(image_path):
        print(f"Error: Image not found at {image_path}")
        return
    if not os.path.exists(label_path):
        print(f"Error: Label not found at {label_path}")
        return
    
    # Загружаем изображение
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Failed to load image at {image_path}")
        return
        
    h, w = image.shape[:2]
    
    # Цвета и метки классов
    class_colors = {
        0: (0, 255, 0),    # Зеленый - person
        1: (0, 0, 255),    # Красный - go_cart
    }
    class_names = {
        0: "person",
        1: "go_cart"
    }
    
    # Читаем аннотации
    try:
        with open(label_path, 'r') as f:
            for line in f.readlines():
                parts = line.split()
                if len(parts) < 5:
                    continue  # Пропускаем невалидные строки
                    
                # Парсим данные
                class_id = int(parts[0])  # Преобразуем в int
                xc, yc, bw, bh = map(float, parts[1:5])
                
                # Рассчитываем координаты
                x1 = int((xc - bw/2) * w)
                y1 = int((yc - bh/2) * h)
                x2 = int((xc + bw/2) * w)
                y2 = int((yc + bh/2) * h)
                
                # Получаем цвет и имя класса
                color = class_colors.get(class_id, (255, 255, 255))  # Белый по умолчанию
                name = class_names.get(class_id, f"class_{class_id}")
                
                # Рисуем bounding box
                cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
                cv2.putText(image, name, (x1, y1-10), 
                            cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
    
    except Exception as e:
        print(f"Error reading label file: {e}")
        return
    
    # Показываем результат
    cv2.imshow("Annotation Check", image)
    key = cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    # Сохраняем результат для последующего просмотра
    output_dir = "annotation_checks"
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, os.path.basename(image_path))
    cv2.imwrite(output_path, image)
    print(f"Saved visualization to {output_path}")


# Пример использования
base_name = "frame_000000"
plot_yolo_boxes(
    f"dataset/train/go_cart_images/{base_name}.jpg",  # Стандартный путь к изображению
    f"dataset/train/go_cart_labels/{base_name}.txt"   # Стандартный путь к аннотации
)
```

```python
from collections import defaultdict

import cv2
import numpy as np

from ultralytics import YOLO

# Load the YOLO11 model
model = YOLO("yolo11n.pt")

# Open the video file
video_path = "Video_go_cart.mp4"
cap = cv2.VideoCapture(video_path)

# Store the track history
track_history = defaultdict(lambda: [])

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLO11 tracking on the frame, persisting tracks between frames
        result = model_cart.track(frame, persist=True, tracker='bytetrack.yaml')[0]

        # Get the boxes and track IDs
        if result.boxes and result.boxes.is_track:
            boxes = result.boxes.xywh.cpu()
            track_ids = result.boxes.id.int().cpu().tolist()

            # Visualize the result on the frame
            frame = result.plot()

            # Plot the tracks
            for box, track_id in zip(boxes, track_ids):
                x, y, w, h = box
                track = track_history[track_id]
                track.append((float(x), float(y)))  # x, y center point
                if len(track) > 30:  # retain 30 tracks for 30 frames
                    track.pop(0)

                # Draw the tracking lines
                points = np.hstack(track).astype(np.int32).reshape((-1, 1, 2))
                cv2.polylines(frame, [points], isClosed=False, color=(0, 230, 0), thickness=10)

        # Display the annotated frame
        cv2.imshow("YOLO11 Tracking", frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
